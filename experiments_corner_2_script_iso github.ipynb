```python
from elecrophy_kernik import *
import myokit
from map_line import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visual(simulation_result, variable, path='results', voltage='Voltage.Vm'):
    if not os.path.isdir(path):
        os.makedirs(path)
    i = 0
    for result in simulation_result:    
        block = result.block2d()
        imgs = block.get2d(voltage)
        for array in imgs:
            plt.imsave(f'{path}/{i}.png', array, cmap='gray', vmin=-100, vmax=50)
            i += 1
```

```python
def corner_simulation(ctags_path='/home/students/myokit_test_ipsc/Myokit_backup 1910 (66 mipt)/new_human_1/ctags2000_0.5.out',
                      conts_path='/home/students/myokit_test_ipsc/Myokit_backup 1910 (66 mipt)/new_human_1/contactM2000_0.5.out',
                      types_path='/home/students/myokit_test_ipsc/Myokit_backup 1910 (66 mipt)/new_human_1/types0.5.out',
                      g_in=1000, 
                      g_gap=500, 
                      g_bound=15, 
                      sizex=250, sizey=250,
                      startx=1250, starty=1750,
                      model='Kernikfull copy.mmt',
                      left_start=235,
                      left_fin=275,
                      down_start=235,
                      down_fin=275,
                      folder_name_down='./results_corner_3',
                      folder_name_left='./results_corner_3',
                      step_size=0.01,
                      stim=1,
                      start_time=20,
                      duration=2,
                      period=350,
                      multiplier=100,
                      time=2500,
                      stim_step=30,
                      voltage='membrane.V'):
    '''
    any path have to be str, it also have to be full
    there is a double protocol: one myoki.protocole after another
    folder_name have to be str
    '''
    model = myokit.load_model(model)
    
    gx, gy  = make_isodomain_map_heter_cut(nx=sizey, ny=sizex, gx=0.8, gy=0.8, gxcn=0.08, gycn=0.08, g_cut=0.001)
    print(np.shape(gx))
    print(np.shape(gy))
    
    freq_list_left=np.arange(left_start, left_fin, 1)
    freq_list_down=np.arange(down_start, down_fin, 1)
    
    for i in range(len(freq_list_left)):
        start_time2=1000+freq_list_left[i]
        peroid2=freq_list_left[i]
        protocol = make_protocol(stim=0.1,
                  start_time=20,
                  duration=2,
                  period=freq_list_left[i],
                  multiplier=24)
        simulation = make_simulation(model, protocol, gx, gy, npixels=(sizex, sizey), 
                                     step_size=step_size)
        for k in range(24):

            result = run_simulation(simulation, protocol, time=time, log_interval=1, 
                                    paced_y=sizey, paced_x=stim_step, start_x=sizex//8,start_y=0,
                                    s1s2=False,
                                    s2_paced_x=sizex//2,
                                    s2_paced_y=sizey//2,
                                    circle=True,
                                    circle_O=(77,25),
                                    circle_R=20)
            visual(result, voltage, path=folder_name_left+f'/circle/{freq_list_left[i]}/left_{freq_list_left[i]}_{k}')

        
    for j in range(len(freq_list_down)):
        start_time2=1000+freq_list_down[j]
        peroid2=freq_list_down[j]
        protocol = make_protocol(stim=0.1,
                  start_time=20,
                  duration=2,
                  period=freq_list_down[j],
                  multiplier=24)
        simulation = make_simulation(model, protocol, gx, gy, npixels=(sizex, sizey), 
                                     step_size=step_size)
        for k in range(24):
            result = run_simulation(simulation, protocol, time=time, log_interval=1, start_x=sizex//4,start_y=0, 
                                    paced_y=sizey//4 - 5, paced_x=sizex//4- 5,
                                    s1s2=False,
                                    s2_paced_x=sizex//2,
                                    s2_paced_y=sizey//2,
                                    circle=True,
                                    circle_O=(77,25),
                                    circle_R=20)
            visual(result, voltage, path=folder_name_down+f'/circle/{freq_list_down[j]}/down_{freq_list_down[j]}_{k}')

    return 'Fin!'
```
